### MLML Seawater Intake



The goal of this exercise is to use principal component analysis to summarize relationships between co-varying variables measured at the MLML seawater intake.

Starter code is provided for loading and plotting the data. The principal component analysis can be conducted using the [kelp example](2-monterey_bay_kelp.ipynb) as a template.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import dates

filename = 'data/mlml_seawater/2016-08.csv'
df = pd.read_csv(filename,parse_dates=[0,1])
```

```python
df.head()
```

```python
all_vars = ['temp','sal','do2','ph','trans','fluor']

ii = np.ones(np.shape(df['temp_flg']),dtype=bool)
for var in all_vars:
    goodi = df[var+'_flg']==1
    ii = ii & goodi

for var in all_vars:
    plt.figure()
    plt.plot(df['pst_time'][ii],df[var][ii])
    plt.title(var)
    dfmt = dates.DateFormatter('%b %d')
    plt.gca().xaxis.set_major_formatter(dfmt)
```
